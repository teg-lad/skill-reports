```python
import pandas as pd
from bokeh.plotting import figure, show, ColumnDataSource
from bokeh.io import output_notebook
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/suzannelittle/ca682i/master/data/sampler/DataSampler.csv")
df.rename(columns={"E":"Fruit"}, inplace=True)
df.describe()
```

```python
output_notebook()
```

```python
tooltips = [
    ("produce", "@Fruit")
]

p = figure(tooltips=tooltips, x_axis_label="A values", y_axis_label="B values")

col = {"banana":"yellow",
       "apple":"red",
       "celery":"lime",
       "orange":"orange",
       "onion":"brown"}


for unique in df.Fruit.unique():
    source = ColumnDataSource(df[df["Fruit"] == unique])
    p.circle("B", "C", legend_label=unique, source=source, fill_color=col[unique], fill_alpha=0.5, size=10, line_color=col[unique])
p.title.text = "Random Beauty of Fruit (and some vegetables)"
p.title.text_color = "purple"

tooltips = [
    ("produce", "$legend")
]

show(p)
```
